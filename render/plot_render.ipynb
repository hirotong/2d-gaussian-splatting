```python
normal = cv2.imread("./render/renderings/gamepad_sony_dualshock_3/front_left_normal0001.png", -1)[..., :3]
normal = normal[..., ::-1]
view = np.load("./render/renderings/gamepad_sony_dualshock_3/front_left.npy")
R = view[:3, :3]
```

```python
normal_show = normal / 255.0
normal = 2 * (normal / 255.0) - 1
normal_view = np.dot(normal, R.T)
normal_view = (normal_view + 1) / 2
# normal_view = normal_view ** (2.2)
```

```python
plt.figure()
ax = plt.subplot(1, 2, 1)
ax.imshow(normal_show)
ax = plt.subplot(1, 2, 2)
ax.imshow(normal_view)
plt.show()
```
