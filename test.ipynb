```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
randn = np.random.randn(100, 3)
rand = np.random.rand(100, 3) * 2 - 1

randn = randn / np.linalg.norm(randn, axis=1, keepdims=True)
rand = rand / np.linalg.norm(rand, axis=1, keepdims=True)
```

```python
ax = plt.subplot(121)
ax.scatter(randn[:, 0], randn[:, 1])
ax.set_title('randn')
ax.axis("equal")
ax = plt.subplot(122)
ax.scatter(rand[:, 0], rand[:, 1])
ax.set_title('rand')
ax.axis("equal")
plt.tight_layout()
```
